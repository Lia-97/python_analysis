```python
# (1)
import pandas as pd

# k:v 구조를 갖는 딕셔너리를 만들고, 변수 dict_data에 저장
dict_data = {0:10,1:20,2:30,3:40,4:50}

# 판다스 Series() 함수로 딕셔너리(dict_data)를 시리즈로 변환. 변수 sr에 저장 
sr = pd.Series(dict_data)

# 변수 sr의 자료형 출력
print(type(sr))

# 변수 sr에 저장되어 있는 시리즈 객체를 출력
print(sr)
print('\n')

#(1)
import pandas as pd

#리스트를 시리즈로 변환하여 변수 sr에 저장
list_data=[10,20,30,40,50]
sr=pd.Series(list_data)
print(type(sr))
print(sr)
```

```python
# (4)
import pandas as pd

#리스트를 시리즈로 변환(인덱스 옵션 지정)
list_data=['둘리','또치','도우너','희동이','마이콜']
sr = pd.Series(list_data,index=['학생1','학생2','학생3','힉생4','학생5'])


print(type(sr))
print(sr)
```

```python
# (5)
blood = ['A형', 'B형', 'O형', 'AB형']
st = [34.2, 27.1, 26.7, 11.5]
sr = pd.Series(st, index=blood)

print('[시리즈 sr 출력]')
print(sr)

print('[시리즈 sr에서 인덱싱과 슬라이싱하여 출력(숫자인덱스 사용)]')
print(sr[0])
print(sr[-1])
print(sr[1])
print(sr[1:4])

print('[시리즈 sr에서 인덱싱과 슬라이싱하여 출력(이름인덱스 사용)]')
print(sr['A형'])
print(sr['AB형'])
print(sr['B형'])
print(sr["B형":"AB형"])

print('[시리즈 sr의 크기 출력]')
print(len(sr))

print('[시리즈 sr의 인덱스 출력]')
print(sr.index)

print('[시리즈 sr의 데이터 값 출력]')
print(sr.values)
```

```python
# (6)
import pandas as pd

# 열이름을 key로 하고, 리스트를 value로 갖는 딕셔너리 정의(2차원 배열)
dict_data = {'name':['둘리','또치','도우너','희동이'],'kor':[90,80,70,70],'eng':[99,98,97,46],'mat':[90,70,70,60]}

# 판다스 DataFrame() 함수로 딕셔너리를 데이터프레임으로 변환. 변수 df에 저장
df = pd.DataFrame(dict_data)

print(type(df))
print(df)
```

```python
# (7)
import pandas as pd

#행 인덱스/열 이름 지정하여 데이터프레임 만들기
df = pd.DataFrame([['둘리','또치','도우너','희동이'],[90,80,70,70],[99,98,97,46],[90,70,70,60]],
                 index=['학생1','학생2','학생3','학생4'],
                 columns=['name','kor','eng','mat'])

print(type(df))
print(df)
print('\n')

#(8)
df.index = ['st1','st2','st3','st4']
df.columns = ['dooly','doochi','dounar','heedong']
print(df)
```

```python
data = {
    'name':['듀크1', '듀크2', '듀크3', '듀크4', '듀크5', '듀크6', '듀크7'],
    'kor':[90, 80, 70, 70, 60, 70, 90],
    'eng':[99, 98, 97, 46, 77, 56, 90],
    'mat':[90, 70, 70, 60, 88, 99, 90],
}
df = pd.DataFrame(data)

print(df)
print('\n')
print(df[0:5])
print(df.head(5))
print('\n')
print(df.tail(5))
print('\n')
print(df.columns)
print('\n')
print(df['name'])
print('\n')
print(df.eng)
print('\n')
print(df[['kor','mat']])
print('\n')
```
